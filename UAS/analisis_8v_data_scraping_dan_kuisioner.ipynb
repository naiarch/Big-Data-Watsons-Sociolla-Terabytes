# **Merge Data**

```python
# 1. Standarisasi kolom Watsons
df_w_std = df_w[['nama_produk', 'harga', 'harga_original', 'diskon', 'rating', 'jumlah_review']].copy()
df_w_std['sumber'] = 'Watsons'
df_w_std['jumlah_terjual'] = None

# 2. Standarisasi kolom Sociolla
df_s_std = df_s[['nama_produk', 'harga', 'harga_original', 'rating', 'jumlah_review', 'jumlah_terjual']].copy()
df_s_std['sumber'] = 'Sociolla'
df_s_std['harga'] = df_s_std['harga_original']
df_s_std['diskon'] = None

# 3. Samakan urutan kolom
cols = ['nama_produk', 'harga', 'harga_original', 'diskon', 'rating', 'jumlah_review', 'jumlah_terjual', 'sumber']
df_w_std = df_w_std[cols]
df_s_std = df_s_std[cols]

# 4. Merge (union / concat)
df_merged = pd.concat([df_w_std, df_s_std], ignore_index=True)

# 5. Preview hasil merge
print("HASIL MERGE DATA")
print(f"Total baris gabungan : {len(df_merged)}")
print(f"  - Watsons          : {len(df_w_std)}")
print(f"  - Sociolla         : {len(df_s_std)}")
print(f"\nKolom               : {list(df_merged.columns)}")
print(f"\nMissing values per kolom:")
print(df_merged.isnull().sum())
print()
df_merged.head(10)
```

Berdasarkan output yang dihasilkan, bahwa dataset hasil merge memiliki delapan variabel, yaitu nama_produk, harga, harga_original, diskon, rating, jumlah_review, jumlah_terjual, dan sumber. Kolom sumber berfungsi sebagai penanda asal data sehingga setiap produk dapat diidentifikasi berasal dari Watsons atau Sociolla. Pada hasil merge terlihat bahwa beberapa nilai masih berupa None atau NaN, khususnya pada kolom jumlah_terjual dan rating. Nilai None pada kolom jumlah_terjual muncul pada data Watsons karena platform  tidak menyediakan informasi jumlah produk yang terjual. Sementara itu, nilai NaN pada kolom rating menunjukkan bahwa terdapat beberapa produk yang belum memiliki penilaian dari pengguna sehingga data rating tidak tersedia.

```python
# V4 - VERACITY: ANALISIS LANJUTAN OUTLIER HARGA
def analisis_outlier_detail(df, harga_col, label, color):
    print(f"  - ANALISIS OUTLIER HARGA — {label.upper()}")

    harga = df[harga_col].dropna()

    # Hitung batas IQR
    q1  = harga.quantile(0.25)
    q3  = harga.quantile(0.75)
    iqr = q3 - q1
    batas_bawah = q1 - 1.5 * iqr
    batas_atas  = q3 + 1.5 * iqr

    print(f"  Q1 (batas bawah normal) : Rp{q1:>12,.0f}")
    print(f"  Q3 (batas atas normal)  : Rp{q3:>12,.0f}")
    print(f"  IQR                     : Rp{iqr:>12,.0f}")
    print(f"  Batas wajar bawah       : Rp{batas_bawah:>12,.0f}")
    print(f"  Batas wajar atas        : Rp{batas_atas:>12,.0f}")

    # Pisahkan outlier atas dan bawah
    outlier_atas  = df[df[harga_col] > batas_atas].copy()
    outlier_bawah = df[df[harga_col] < batas_bawah].copy()

    print(f"\n  - Outlier harga TERLALU MAHAL : {len(outlier_atas)} produk")
    print(f"  - Outlier harga TERLALU MURAH : {len(outlier_bawah)} produk")

    # Visualisasi
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))
    fig.suptitle(f'Analisis Outlier Harga — {label}', fontsize=14, fontweight='bold')

    # Boxplot
    axes[0].boxplot(harga, patch_artist=True,
                    boxprops=dict(facecolor=color, alpha=0.6, color=dark),
                    medianprops=dict(color=accent, linewidth=2.5),
                    whiskerprops=dict(color=dark), capprops=dict(color=dark),
                    flierprops=dict(marker='o', color='red', alpha=0.5, markersize=5))
    axes[0].axhline(batas_atas,  color='red',   linestyle='--', linewidth=1.5, label=f'Batas atas Rp{batas_atas:,.0f}')
    axes[0].axhline(batas_bawah, color='orange', linestyle='--', linewidth=1.5, label=f'Batas bawah Rp{batas_bawah:,.0f}')
    axes[0].set_title('Boxplot Harga', fontsize=12, fontweight='bold')
    axes[0].set_ylabel('Harga (Rp)')
    axes[0].legend(fontsize=8)

    # Scatter: semua produk, outlier diberi warna berbeda
    normal      = df[(df[harga_col] >= batas_bawah) & (df[harga_col] <= batas_atas)]
    axes[1].scatter(range(len(normal)),      normal[harga_col],
                    color=color,   alpha=0.3, s=15, label=f'Normal ({len(normal)})')
    axes[1].scatter(range(len(outlier_atas)), outlier_atas[harga_col],
                    color='red',   alpha=0.7, s=40, label=f'Outlier Mahal ({len(outlier_atas)})', zorder=5)
    if len(outlier_bawah) > 0:
        axes[1].scatter(range(len(outlier_bawah)), outlier_bawah[harga_col],
                        color='orange', alpha=0.7, s=40, label=f'Outlier Murah ({len(outlier_bawah)})', zorder=5)
    axes[1].axhline(batas_atas,  color='red',   linestyle='--', linewidth=1.2)
    axes[1].axhline(batas_bawah, color='orange', linestyle='--', linewidth=1.2)
    axes[1].set_title('Sebaran Harga Semua Produk', fontsize=12, fontweight='bold')
    axes[1].set_ylabel('Harga (Rp)')
    axes[1].set_xlabel('Index Produk')
    axes[1].legend(fontsize=8)

    # Bar: top 10 outlier termahal
    top10_mahal = outlier_atas.nlargest(10, harga_col).sort_values(harga_col, ascending=True)
    if len(top10_mahal) > 0:
        axes[2].barh(top10_mahal['nama_produk'].str[:35],
                     top10_mahal[harga_col],
                     color='red', alpha=0.7, edgecolor=dark)
        axes[2].axvline(batas_atas, color='gray', linestyle='--',
                        linewidth=1.5, label=f'Batas wajar Rp{batas_atas:,.0f}')
        axes[2].set_title('Top 10 Produk Outlier Termahal', fontsize=12, fontweight='bold')
        axes[2].set_xlabel('Harga (Rp)')
        axes[2].legend(fontsize=8)

    plt.tight_layout()
    plt.show()

    # Tabel Rincian Outlier Mahal
    print(f"\n  - RINCIAN OUTLIER HARGA MAHAL — {label}")
    print(f"  (Produk dengan harga > Rp{batas_atas:,.0f})\n")

    kolom_tampil = ['nama_produk', harga_col, 'rating', 'jumlah_review']
    kolom_tampil = [k for k in kolom_tampil if k in df.columns]

    tampil_atas = outlier_atas[kolom_tampil].sort_values(harga_col, ascending=False).head(15)
    tampil_atas[harga_col] = tampil_atas[harga_col].apply(lambda x: f'Rp{x:,.0f}')
    print(tampil_atas.to_string(index=False))

    # Tabel Rincian Outlier Murah
    if len(outlier_bawah) > 0:
        print(f"\n  - RINCIAN OUTLIER HARGA MURAH — {label}")
        print(f"  (Produk dengan harga < Rp{batas_bawah:,.0f})\n")
        tampil_bawah = outlier_bawah[kolom_tampil].sort_values(harga_col).head(15)
        tampil_bawah[harga_col] = tampil_bawah[harga_col].apply(lambda x: f'Rp{x:,.0f}')
        print(tampil_bawah.to_string(index=False))
    else:
        print(f"\n  - Tidak ada outlier harga terlalu murah di {label}.")

    # Flag otomatis: mana yang perlu ditangani
    print(f"  - PRODUK YANG PERLU DICEK ULANG (outlier mahal + rating < 3.0 ATAU review = 0):\n")
    suspect = outlier_atas[
        (outlier_atas['rating'] < 3.0) |
        (outlier_atas['jumlah_review'] == 0) |
        (outlier_atas['rating'].isna())
    ][kolom_tampil].sort_values(harga_col, ascending=False)

    if len(suspect) > 0:
        suspect_tampil = suspect.copy()
        suspect_tampil[harga_col] = suspect_tampil[harga_col].apply(lambda x: f'Rp{x:,.0f}')
        print(suspect_tampil.to_string(index=False))
        print(f"\n  - Total {len(suspect)} produk patut dicurigai sebagai data error.")
    else:
        print(f"  - Semua outlier mahal memiliki rating & review yang wajar.")
        print(f"     Kemungkinan besar ini memang produk premium — TIDAK perlu ditangani.")

    return outlier_atas, outlier_bawah, batas_bawah, batas_atas


# Jalankan untuk Watsons
outlier_atas_w, outlier_bawah_w, batas_bawah_w, batas_atas_w = analisis_outlier_detail(
    df_w_std, 'harga', 'Watsons', watsons_color
)

# Jalankan untuk Sociolla
outlier_atas_s, outlier_bawah_s, batas_bawah_s, batas_atas_s = analisis_outlier_detail(
    df_s_std, 'harga_original', 'Sociolla', sociolla_color
)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns
import numpy as np
from scipy.stats import chi2_contingency

# Rename kolom responden
df_r.columns = [
    'timestamp', 'nama', 'gender', 'usia', 'spending', 'toko_digunakan',
    'w_kelengkapan', 'w_miss', 'w_review', 'w_diskon', 'w_rating', 'col11',
    'w_terjual', 'w_kategori',
    's_kelengkapan', 's_miss', 's_review', 's_diskon', 's_rating', 's_terjual',
    's_kategori', 'rekomendasi', 'score', 'email'
]

# Tema warna
W_COLOR = '#3399FF'
S_COLOR = '#FF007F'
NEUTRAL = '#FFD700'

print('Data berhasil dimuat!')
print(f'   Watsons  : {len(df_w):,} produk')
print(f'   Sociolla : {len(df_s):,} produk')
print(f'   Responden: {len(df_r)} orang')
```

```python
# Normalisasi metrik scraping ke skala 1–4 agar apple-to-apple dengan Likert
# Rating: scraping (1–5) → skala 1–4
w_rating_norm = (df_w['rating'].mean() - 1) / (5 - 1) * 3 + 1
s_rating_norm = (df_s['rating'].mean() - 1) / (5 - 1) * 3 + 1

# Diskon: persen rata-rata → skala 1–4 (max reference = 50%)
w_diskon_norm = 1 + (df_w['diskon'].mean() / 50) * 3
s_diskon_norm = 1.0  # Sociolla tidak mencatat diskon di data scraping

# Review: rata-rata jumlah review → normalisasi log scale ke 1–4
w_rev_log   = np.log1p(df_w['jumlah_review'].mean())
s_rev_log   = np.log1p(df_s['jumlah_review'].mean())
max_log     = max(w_rev_log, s_rev_log)
w_rev_norm  = 1 + (w_rev_log / max_log) * 3
s_rev_norm  = 1 + (s_rev_log / max_log) * 3

# Jumlah terjual Sociolla → normalisasi log
s_terjual_norm = 1 + (np.log1p(df_s['jumlah_terjual'].mean()) / max_log) * 3

# Persepsi rata-rata dari responden
dimensi_label = ['Rating\nProduk', 'Pengaruh\nDiskon', 'Frekuensi\nNulis Review', 'Pengaruh\nJml Terjual']
persepsi_w    = [df_r['w_rating'].mean(), df_r['w_diskon'].mean(), df_r['w_review'].mean(), df_r['w_terjual'].mean()]
persepsi_s    = [df_r['s_rating'].mean(), df_r['s_diskon'].mean(), df_r['s_review'].mean(), df_r['s_terjual'].mean()]
aktual_w      = [w_rating_norm, w_diskon_norm, w_rev_norm, 2.5]  # Watsons tdk punya data jumlah_terjual
aktual_s      = [s_rating_norm, s_diskon_norm, s_rev_norm, s_terjual_norm]

x = np.arange(len(dimensi_label))
width = 0.18

fig, axes = plt.subplots(1, 2, figsize=(16, 6), sharey=True)
fig.suptitle('GAP ANALYSIS: Persepsi Responden vs Data Scraping (Skala 1–4)',
             fontsize=15, fontweight='bold', y=1.02)

for ax, (toko, p_vals, a_vals, color) in zip(axes, [
    ('Watsons',  persepsi_w, aktual_w, W_COLOR),
    ('Sociolla', persepsi_s, aktual_s, S_COLOR)
]):
    b1 = ax.bar(x - width/2, p_vals, width, label='Persepsi Responden',
                color=color, alpha=0.85, edgecolor='white')
    b2 = ax.bar(x + width/2, a_vals, width, label='Data Aktual (Scraping)',
                color=color, alpha=0.35, edgecolor=color, linewidth=1.5)
    for i, (p, a) in enumerate(zip(p_vals, a_vals)):
        gap = p - a
        clr = '#e74c3c' if gap > 0.3 else '#27ae60' if gap < -0.3 else 'gray'
        ax.annotate(f'Δ{gap:+.2f}', xy=(x[i], max(p, a) + 0.05),
                    ha='center', fontsize=9, color=clr, fontweight='bold')
    ax.set_title(toko, fontsize=13, fontweight='bold', color=color)
    ax.set_xticks(x)
    ax.set_xticklabels(dimensi_label, fontsize=9)
    ax.set_ylim(0, 4.8)
    ax.set_ylabel('Skor (skala 1–4)', fontsize=10)
    ax.legend(fontsize=9)
    ax.grid(axis='y', alpha=0.3, linestyle='--')
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.savefig('gap_analysis.png', dpi=150, bbox_inches='tight')
plt.show()

print('\nINSIGHT:')
print(f'   • Diskon Watsons: responden sangat terpengaruh (skor {df_r["w_diskon"].mean():.2f}/4),')
print(f'     terkonfirmasi oleh data scraping: rata-rata diskon = {df_w["diskon"].mean():.1f}%')
print(f'   • Rating Watsons aktual sangat tinggi ({df_w["rating"].mean():.2f}/5) namun')
print(f'     persepsi responden hanya {df_r["w_rating"].mean():.2f}/4 → responden underestimate')
print(f'   • Review Sociolla jauh lebih banyak secara aktual ({df_s["jumlah_review"].mean():.0f}/produk)')
print(f'     vs Watsons ({df_w["jumlah_review"].mean():.0f}/produk)')
```

Berdasarkan analisis  gabungan data kuesioner dan data *scraping*, menunjukkann bahwa terdapat pola perilaku serta persepsi konsumen antara pengguna Sociolla dan Watsons. Dari sisi demografi, mayoritas konsumen perempuan menunjukkan kecenderungan kuat untuk memilih Sociolla, sedangkan sebagian besar konsumen laki-laki lebih memilih Watsons. Selain faktor gender, daya beli juga turut membentuk segmentasi pasar ini, di mana konsumen dengan tingkat pengeluaran tinggi cenderung memilih Sociolla, yang sekaligus mempertegas posisi platform tersebut pada segmen pasar yang lebih premium.

Kecenderungan pilihan pada Sociolla dinilai oleh konsumen lebih unggul dalam hal kelengkapan produk yang disediakan. Di sisi lain, Watsons memiliki daya tarik yang jauh lebih kuat dalam hal pengaruh promosi, di mana konsumen menganggap diskon yang ditawarkan Watsons sangat memikat. Sementara itu, rating toko dan jumlah produk terjual dinilai memiliki pengaruh yang relatif seimbang dalam memengaruhi keputusan belanja di kedua platform tersebut.

Analisis kesenjangan atau *gap analysis*, menunjukkan bahwa persepsi konsumen mengenai kuatnya daya tarik diskon di Watsons adalah hal yang akurat, karena data *scraping* membuktikan rata-rata diskon di platform tersebut mencapai angka belasan persen. Namun, terjadi fenomena di mana konsumen menilai rendah kualitas Watsons secara umum, padahal data objektif hasil *scraping* menunjukkan bahwa rating toko Watsons sangat tinggi hingga hampir menyentuh angka sempurna. Terakhir, ditemukan pula adanya celah perilaku yang besar terkait interaksi pasca-pembelian, di mana lebih dari separuh responden kuesioner mengaku tidak pernah menulis ulasan setelah berbelanja, baik di Sociolla maupun di Watsons.
