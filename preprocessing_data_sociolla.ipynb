```python
# @title drop kolom
df = df.drop(columns=['brand_name', 'product_id', 'price_by_combinations', 'beauty_point_earned', 'url', 'active_date', 'default_category', 'categories', 'rating_types_str', 'average_rating_by_types'])
df
```

```python
df.info()
```

# **Adjusting Column**

```python
# @title kolom harga
def clean_and_average_price(price_str):
    cleaned = re.sub(r'[^\d\-]', '', str(price_str))
    if not cleaned:
        return 0.0
    if '-' in cleaned:
        prices = cleaned.split('-')
        valid_prices = [float(p) for p in prices if p]
        return sum(valid_prices) / len(valid_prices) if valid_prices else 0.0
    else:
        return float(cleaned)

df['harga'] = df['price_range'].apply(clean_and_average_price)
df = df.drop(columns=['price_range'])
df
```

```python
# @title kolom jumlah_terjual
df['jumlah_terjual'] = (
    df['total_repurchase_maybe_count'] +
    df['total_repurchase_no_count'] +
    df['total_repurchase_yes_count']
)

kolom_yang_didrop = ['total_repurchase_maybe_count', 'total_repurchase_no_count',
                     'total_repurchase_yes_count']
df = df.drop(columns=kolom_yang_didrop)
df
```

```python
# @title fixed column
df['toko'] = 'Sociolla'

kolom_baru = {
    'product_name': 'nama_produk',
    'average_rating': 'rating',
    'total_reviews': 'jumlah_review',
    'total_recommended_count': 'jumlah_direkomendasikan',
    'total_in_wishlist': 'jumlah_wishlist'
}
df = df.rename(columns=kolom_baru)

urutan_kolom = [
    'toko',
    'nama_produk',
    'harga',
    'rating',
    'jumlah_review',
    'jumlah_terjual',
    'jumlah_direkomendasikan',
    'jumlah_wishlist'
]
df = df[urutan_kolom]

df
```

# **Menyimpan Dataset Sociolla**

```python
df.to_csv('products_sociolla_clean.csv', index=False)
print("Dataset berhasil disimpan dengan nama 'products_sociolla_clean.csv'!")
```
