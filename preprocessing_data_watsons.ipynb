# Load Data

```python
url = "https://raw.githubusercontent.com/naiarch/UTS-Big-Data-Web-Scraping-Lazada/main/product_watsons.csv"

df = pd.read_csv(url)
df.head()
```

# Cleaning Data

```python
# @title drop kolom
df = df.drop(columns=['kategori', 'harga_member', 'lokasi', 'brand', 'url_produk', 'url_gambar', 'tanggal_scraping', 'halaman'])
df
```

```python
df.info()
```

```python
# @title cleaning harga

def clean_price(price):
    if pd.isna(price): return np.nan
    cleaned = (str(price).replace('Rp', '').replace('.', '').replace(',', '').strip())
    return float(cleaned)

df['harga'] = df['harga'].apply(clean_price)
df['harga_original'] = df['harga_original'].fillna(df['harga'])
df
```

```python
# @title cleaning diskon

def clean_discount(diskon):
    cleaned = re.sub(r'[^\d.]', '', str(diskon))
    if cleaned == '': return 0.0
    return float(cleaned)

df['diskon'] = df['diskon'].apply(clean_discount)
df
```

```python
# @title kolom jumlah_review
df['jumlah_review'] = pd.to_numeric(df['jumlah_review'], errors='coerce')
df['jumlah_review'] = df['jumlah_review'].fillna(0)
df
```

```python
# @title kolom jumlah_terjual

# jumlah_terjual tidak diimputasi karena data memang tidak tersedia dari platform
df['jumlah_terjual'] = pd.to_numeric(df['jumlah_terjual'], errors='coerce')
df
```

```python
# @title cleaning rating

# mengubah rating menjadi numerik
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df
```

```python
# @title fixed column

# mengatur ulang urutan kolom
urutan_kolom = [
    'toko',
    'nama_produk',
    'harga',
    'harga_original',
    'diskon',
    'rating',
    'jumlah_review',
    'jumlah_terjual'
]

df = df[urutan_kolom]
df
```
